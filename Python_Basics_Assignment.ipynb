 6 .    Why are lists mutable, but tuples are immutable ?
             List are mutable :
            Lists are designed to be flexible and dynamic.They store references to objects in a contiguous block of memory that can change in size — items can be added, removed, or replaced.
          
             Tuples are immutable :
             Tuples are meant to be fixed collections of items - like records or constants.Once created, the memory layout of a tuple is fixed, so it cannot be resized or altered.

13 . What is the difference between for and while loops ?
           for loops :
           =>Used when you know in advance how many times you want to repeat   something.
           =>It iterates over a sequence (like a list, string, or range).
           while loops :
            => Used when you don’t know beforehand how many times to loop.
            => It keeps running as long as a condition is True.

14 . Describe a scenario where a while loop is more suitable than a for loop ?
           Suppose you want to repeatedly ask the user to enter a password until they get it right.
          You don’t know how many attempts they will need — it could be 1 time or 100 times.
          Why a while loop is better here:

         The number of repetitions is unknown.
         You only stop when a condition (user_input == correct_password) becomes true.
          A for loop would not make sense because you’d have to set a fixed number of attempts.

```python
(10)# Write code to demonstrate type casting with list elements.

# List with mixed data types
values = ["10", "20", "30", "40"]

print("Original list:", values)
print("Data type of first element:", type(values[0]))

# Type casting each element from string to integer
for i in range(len(values)):
    values[i] = int(values[i])

print("\nList after type casting:", values)
print("Data type of first element now:", type(values[0]))

```

```python
(14)# Write a program to reverse a string using a while loop.
string = "hello"
reversed_string = ""

i = len(string) - 1  # Start from the last index

while i >= 0:
    reversed_string += string[i]
    i -= 1

print("Reversed string:", reversed_string)
```

```python
(15)# Write a Python program to calculate the factorial of a number provided by the user using a while loop?

num = int(input("Enter a number: "))
factorial = 1

i = 1
while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is", factorial)
```
